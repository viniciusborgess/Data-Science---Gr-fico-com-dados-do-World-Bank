# 1. Importando Bibliotecas Necessárias.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 2. Carregando os Dados CSV no DataFrame Pandas.

(Loading CSV Data into Pandas DataFrame)

```python
df = pd.read_csv('dataset.csv', encoding='utf-8')

df
# Informações gerais sobre o dataset
df.info()
# Estatísticas descritivas
df.describe()
```

# 3. Análise e Visualização dos Dados

Gráfico de Linha usado para mostrar a tendência do PIB ao longo dos anos.

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Ano'], df['PIB'], marker='o')
plt.title('PIB ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('PIB (em bilhões)')
plt.grid(True)
plt.show()

```

Gráfico de barras usado para comparar a população ao longo dos anos.

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Ano'], df['População'], color='orange')
plt.title('População ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('População (em milhões)')
plt.grid(True)
plt.show()
```

Gráfico de histograma usado para mostrar a distribuição da taxa de desemprego ao longo dos anos.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Taxa_desemprego'], bins=5, color='green')
plt.title('Distribuição da Taxa de Desemprego')
plt.xlabel('Taxa de Desemprego (%)')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

Gráfico de pizza usado para mostrar a proporção das emissões de CO2 por setor no ano de 2020.

```python
setores = ['Indústria', 'Transporte', 'Agricultura', 'Residencial', 'Comercial']
emissoes = [150, 120, 70, 40, 20]

plt.figure(figsize=(8, 8))
plt.pie(emissoes, labels=setores, autopct='%1.1f%%', startangle=140)
plt.title('Emissão de CO2 por setor em 2020')
plt.show()
```

Gráfico de Scatter usado para mostrar a relação entre o consumo de energia e a emissão de CO2.


```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Consumo_energia'], df['Emissao_CO2'], color='red')
plt.title('Consumo de Energia vs Emissão de CO2')
plt.xlabel('Consumo de Energia (em TWh)')
plt.ylabel('Emissão de CO2 (em milhões de toneladas)')
plt.grid(True)
plt.show()
```

### Explicação dos Gráficos

* **Linha**: Usado para mostrar a tendência do PIB ao longo dos anos.
* **Barra**: Usado para comparar a população ao longo dos anos.
* **Histograma**: Usado para mostrar a distribuição da taxa de desemprego ao longo dos anos.
* **Pizza**: Usado para mostrar a proporção das emissões de CO2 por setor no ano de 2020.
* **Scatter**: Usado para mostrar a relação entre o consumo de energia e a emissão de CO2.
